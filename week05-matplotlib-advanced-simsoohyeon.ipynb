```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)

plt.plot([1, 2, 3, 4], [10,20,30,40])
plt.title('plotting')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(7,4))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

x = np.linspace(-2*np.pi, 2*np.pi, 200)
plt.plot(x, np.sin(x)*10)
plt.plot(x, x**2)
plt.plot(x, 2*x**2-3*x)

plt.grid(True)
plt.title('다양한 그래프')
plt.show()
```

## plt.subplot()

```python
ax1=plt.subplot(2,1,1)
plt.plot(np.random.randint(1,10,10), 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2=plt.subplot(2,1,2)
plt.plot(np.arange(10), np.arange(10)*2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
fig, axes=plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title("axes 1")
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title("axes 2")
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title("axes 3")
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,2)

axes[0,0].plot([1,2,3,4])
axes[0,0].set_title("Subplot 1")

axes[0,1].plot([4,3,2,1])
axes[0,1].set_title("Subplot 2")

axes[1,0].hist([1,2,2,3,3,3,4,4,4,4,4], bins=5)
axes[1,0].set_title("Subplot 3")

axes[1,1].scatter([1,2,3,4], [4,3,2,1])
axes[1,1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

## fig.add_subplot()

```python
fig = plt.figure(figsize=(10,8))

ax2=fig.add_subplot(2,2,2)
plt.plot(np.random.randn(50).cumsum(), 'k--')
ax3=fig.add_subplot(2,2,3)
```
